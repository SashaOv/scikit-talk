##Read Data

```python
fmli= pd.read_csv("data/fmli121x.csv")
fmli.describe()
```

##Project

```python
src_col=['NEWID','FAM_SIZE','NO_EARNR','AGE_REF','EDUC_REF','STATE','PSU',
         'FINCBTAX', 'TOTTXPDM','FINDRETX','PERINSPQ',
         'TOTEXPPQ','HOUSPQ','APPARPQ','TRANSPQ','HEALTHPQ','ENTERTPQ',
         'SMSASTAT','VEHQ','SWIMPOOL']
dest_col=['id','FAM_SIZE','NO_EARNR','AGE_REF','EDUC_REF','STATE','metro',
          'income', 'taxes','savings','insurance',
          'expenses','housing','apparel','transport','health','entertainment',
          'metropolitan','cars','SWIMPOOL']
ds=fmli[src_col]
ds.columns=dest_col

ds['food']=fmli['FOODPQ'] + fmli['ALCBEVPQ']
ds['other']=fmli['PERSCAPQ'] + fmli['READPQ'] + fmli['EDUCAPQ'] + \
    fmli['TOBACCPQ'] + fmli['MISCPQ'] + fmli['CASHCOPQ'] 
ds['own']=(fmli['CUTENURE'] <= 3) + 0                                                                                  
ds.describe()
```

## Clean up

```python
ds= ds[(ds['income'] > 0) & (ds['income'] < 400000) & 
        (ds['taxes'] < ds['income']) & (ds['expenses'] < ds['income'])]
ds.describe()
```

##Curve Fitting

```python
import sklearn.linear_model as lm
import sklearn.cross_validation as cv

x= ds['income'].values
y= ds['taxes'].values

(x_train, x_test, y_train, y_test) = cv.train_test_split(x, y, test_size=.25)
lrp = lm.LinearRegression()
deg= 1
# Generate Vandermonde matrix
lrp.fit(np.vander(x_train, deg + 1), y_train)
y_fit= lrp.predict(np.vander(x_test, deg + 1))
plt.plot(x_train,y_train,'.',label='data')
plt.plot(x_test,y_fit,'or',label='fit')
plt.legend()
```

```python
lrp.score(np.vander(x_test, deg + 1),y_test)
```

```python
import sklearn.ensemble as ske
reg = ske.RandomForestRegressor()
```

```python
vars= ['own','income','taxes','EDUC_REF','AGE_REF','STATE','metro','metropolitan',
       'expenses','housing','apparel','transport','health','entertainment']
cleaned= ds[vars].dropna()

X=np.delete(cleaned.values.copy(),0,1)
y=cleaned['own'].values

reg.fit(X,y)
reg.feature_importances_
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(ds['AGE_REF'],ds['income'],c=ds['own'])
```

```python
indexes= np.argsort(reg.feature_importances_)[::-1].astype(int)
features= pd.DataFrame(reg.feature_importances_[indexes],columns=["importance"],
                       index=np.array(vars[1:])[indexes])
features.plot(kind='bar')
```

```python
owners=ds[['own','income']]
owners= owners.dropna()
X=np.delete(owners.values.copy(),0,1)
y=owners['own'].values
```

```python
(X_train, X_test, y_train, y_test) = cv.train_test_split(X, y, test_size=.2)
logreg = lm.LogisticRegression()
logreg.fit(X_train, y_train)
y_predicted = logreg.predict(X_test)
logreg.score(X,y)
```

```python
cv.cross_val_score(logreg, X, y)
```
